# Creating a dataframe from an array

```python
# Importinglibraries
import pandas as pd
import numpy as np
```

```python
# creating an array
data = np.array([[1, 4], [2, 5], [3, 6]])
```

```python
data
```

```python
# creating a dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'],
                 columns=['col1', 'col2'])
```

```python
# showing the datafrane
df
```

# Creating a DataFrame from a dictionary

```python
# Importinglibraries
import pandas as pd

# Creating a DataFrame from a dictionary
#example
data1 = {'Name': ['Mohammed', 'ALi', 'Malak', 'Reem'],
        'Age': [25, 39, 35, 30],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago'],
        'Height': [177, 180, 165, 155]}
```

```python
data1
```

```python
df1 = pd.DataFrame(data1)
df1
```
